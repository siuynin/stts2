# 🚀 StyleTTS2-Lite Vietnamese Fine-tuning (V7 - Sửa lỗi đuôi file Config)

Notebook này đảm bảo mọi script đều tìm thấy file cấu hình dù là .yml hay .yaml.

```python
import os, sys, platform, urllib.request, yaml

os.makedirs("Models/Finetune", exist_ok=True)
os.makedirs("Configs", exist_ok=True)

print("📥 Đang tải Model và Config...")
model_url = "https://huggingface.co/dangtr0408/StyleTTS2-lite-vi/resolve/main/Models/base_model_120k_vi.pth"
config_url = "https://huggingface.co/dangtr0408/StyleTTS2-lite-vi/resolve/main/Models/config.yaml"

if not os.path.exists("Models/Finetune/current_model.pth"):
    urllib.request.urlretrieve(model_url, "Models/Finetune/current_model.pth")

# Lưu thành cả hai đuôi .yml và .yaml để tránh lỗi File Not Found
urllib.request.urlretrieve(config_url, "Configs/config.yml")
urllib.request.urlretrieve(config_url, "Configs/config.yaml")

for cfg_path in ["Configs/config.yml", "Configs/config.yaml"]:
    with open(cfg_path, 'r', encoding='utf-8') as f:
        config = yaml.safe_load(f)
    
    config['pretrained_model'] = './Models/Finetune/current_model.pth'
    config['load_only_params'] = True
    config['data_params']['train_data'] = 'Data/train_list.txt'
    config['data_params']['val_data'] = 'Data/val_list.txt'
    config['data_params']['root_path'] = './Data/'
    
    with open(cfg_path, 'w', encoding='utf-8') as f:
        yaml.dump(config, f)

print("✅ Đã tải và cấu hình xong cả Config.yml và Config.yaml!")
```
